creating a fibonacci seq and plotting it

```python
def generateFibo(a):
    fibonacci_numbers = [0, 1]

    for i in range(2,a):
        fibonacci_numbers.append(fibonacci_numbers[i-1]+fibonacci_numbers[i-2])
    
    return fibonacci_numbers

list1 = generateFibo(25)
list2 = np.random.randint(0,650,25) # 25 random int no from 0 to 650
```

```python
plt.plot(list1,color = 'blue')
plt.plot(list2,color = 'red')
```
